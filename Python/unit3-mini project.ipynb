### Task3: Data Wrangling

```python
df.drop('fips',axis=1,inplace=True)
df.head()
```

#### Rename columns "cases" and "deaths" to "new cases" and "new deaths"

```python
df=df.rename(columns={'cases':'new cases','deaths':'new deaths'})
df
```

```python
df.describe()
```

```python
df.plot(kind="hist")
```

From histogram plot above, we know that the number of new cases are greater than the number of new deaths. New cases in the United States are increasing.
